```python
link2="https://en.wikipedia.org/wiki/List_of_ISO_3166_country_codes"
link3="https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"

import pandas as pd

data2=pd.read_html(link2,header=1,flavor="bs4",attrs={'class':"wikitable sortable"})
data3=pd.read_html(link3,header=0,flavor="bs4",attrs={'class':"wikitable sortable"})
```

```python
isodata=data2[0].copy()
```

```python
isodata.head()
```

```python
isodata.columns=isodata.columns.str.replace("\\s|\\[[0-9]+]","",regex=True)
isodata.columns
```

```python
isodata['Alpha-2code'].str.extract("([A-Z][A-Z])")
```

```python
isodata["iso2"]=isodata['Alpha-2code'].str.extract("([A-Z][A-Z])")
isodata["iso3"]=isodata['Alpha-3code'].str.extract("([A-Z][A-Z][A-Z])")
```

```python
isodata.dropna(subset=["iso2","iso3"],inplace=True)
isodata.reset_index(drop=True,inplace=True)
isodata
```

```python
isodata[isodata.InternetccTLD.str.startswith("[")]
```

```python
isodata[isodata.InternetccTLD.str.match("(.[a-z][a-z]\\s.[a-z][a-z])")]
```

```python
isodata.loc[isodata.InternetccTLD.str.startswith("["),"InternetccTLD"]
```

```python
isodata.loc[isodata.InternetccTLD.str.match("(.[a-z][a-z]\\s.[a-z][a-z])"),"InternetccTLD"]
```

```python
isodata.loc[isodata.InternetccTLD.str.match("(.[a-z][a-z]\\s.[a-z][a-z])"),"InternetccTLD"]=['.nl','.uk']
```

```python
isodata.loc[isodata.InternetccTLD.str.startswith("["),"InternetccTLD"]=None
```

```python
isodata.loc[[30,212,234,244],:]
```

```python
isodata.loc[[27,233],:]
```

```python
whichToDrop=range(2,7)
isodata.drop(labels=isodata.columns[whichToDrop],axis=1,inplace=True)
```

```python
isodata.to_csv("isodata,csv",index=False)
isodata
```

```python
regiondata=data3[0].copy()
regiondata.head()
```

```python
WhichToDrop=list(set(range(8))-set([1,2]))
WhichToDrop
```

```python
regiondata.drop(labels=regiondata.columns[WhichToDrop],axis=1,inplace=True)
regiondata
```

```python
regiondata.Region.value_counts()
```

```python
regiondata=regiondata[regiondata.Region!="World"]
regiondata.reset_index(drop=True,inplace=True)
regiondata
```

```python
regiondata.rename(columns={regiondata.columns.to_list()[0]:"Country"},inplace=True)
```

```python
regiondata[regiondata["Country"].str.endswith("]")]
```

```python
regiondata.Country=regiondata.Country.str.replace("(\\[[a-z]\\])","",regex=True)
```

```python
regiondata[['Country', 'CountryMain']] = regiondata['Country'].str.split('(', 1, expand=True)
```

```python
#regiondata[regiondata.CountryMain]
regiondata[~pd.isna(regiondata.CountryMain)]
```

```python
regiondata.CountryMain=regiondata.CountryMain.str.replace(")","")
```
